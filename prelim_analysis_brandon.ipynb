```python
import pandas as pd
# Makes a dataframe of items on characters and other information
data = {'PlayerName': [],
        'Character': [],
        'Items': [],
        'Place': []}
final_unit_item_combos = pd.DataFrame(data)
for i in range(len(game_data)):
    playerName = game_data[i]['playerName']
    final = game_data[i]['player']['final']
    place = int(final['placement']['rank'])
    for char in final['comp']:
        if (len(char['items']) > 1):
            character = char['character_name']
            items = tuple(char['items'])
            new_row = {'PlayerName': playerName, 
                       'Character': character, 
                       'Items': items, 
                       'Place': place}
            final_unit_item_combos = final_unit_item_combos.append(new_row, ignore_index = True)

print(final_unit_item_combos)
```

```python
# Returns average place of items where times played is greater or equal to times_played
def items_avg_place(times_played):
    item_placements = final_unit_item_combos[final_unit_item_combos['Place'] != 0].groupby(
        ['Items'], as_index = False
    ).agg({'Place':list})
    item_placements['TimesPlayed'] = item_placements['Place'].map(len) 
    more_than_once = item_placements[item_placements['TimesPlayed'] >= times_played]
    item_placements['AvgPlace'] = more_than_once['Place'].map(sum)/more_than_once['TimesPlayed']
    best_avg_place = item_placements.sort_values(by = 'AvgPlace')
    notNaN = best_avg_place['AvgPlace'].notnull()
    return best_avg_place[notNaN]

items_place = items_avg_place(30)
print(items_place)
```

```python
# Count of items combinations on characters
item_combo_dict = {}
for item in final_unit_item_combos['Items']:
    if item not in item_combo_dict:
        item_combo_dict[item] = 1
    else:
        item_combo_dict[item] += 1

sorted_item_combos = sorted(item_combo_dict.items(), key = lambda x: x[1], reverse = True)
print(sorted_item_combos)

    
```

```python
# Takes three item combinations and turns it into unique pairs of item combos
# Returns a list of frozensets
def get_item_pairs(item_tuple):
    item_pairs = set([])
    for i in range(len(item_tuple)):
        for j in range(i + 1, len(item_tuple)):
            pair = (item_tuple[i], item_tuple[j])
            if pair not in item_pairs:
                item_pairs.add(pair)
    return item_pairs

# Count of items pair combinations on characters
item_pair_combo_dict = {}
for item_count_tuple in sorted_item_combos:
    items = item_count_tuple[0]
    count = item_count_tuple[1]
    if len(items) == 3:
        for pair in get_item_pairs(items):
            if pair not in item_pair_combo_dict:
                item_pair_combo_dict[pair] = count
            else:
                item_pair_combo_dict[pair] += count
    else:
        if items not in item_pair_combo_dict:
            item_pair_combo_dict[items] = count
        else:
            item_pair_combo_dict[items] += count

sorted_item_pair_combos = sorted(item_pair_combo_dict.items(), key = lambda x: x[1], reverse = True)
print(sorted_item_pair_combos)
```
