```python
playerNameList = []
roundItemsListSingleP = []
roundItemListWholeP = []
roundVsList = []
for idx in range(0,len(jsonFile)):
    playerNameList.append(jsonFile[idx]['playerName'])
#     print(idx)
    roundVs = list(jsonFile[idx]['player']["round_outcomes"]["by_round"].keys())
    for idx2 in roundVs:
#         print(idx2)
        try:
            roundItem = jsonFile[idx]['player']["round_outcomes"]["by_round"][idx2]["board"][0]["items"]
            roundItemsListSingleP.append(roundItem)
        except IndexError:
            roundItemsListSingleP.append([])
    roundItemListWholeP.append(roundItemsListSingleP)
    roundVsList.append(roundVs)
```

```python
df2 = pd.DataFrame(list(zip(playerNameList,roundVsList,roundItemListWholeP)),columns=["PlayerName","RoundVsList","RoundItems"])
```
